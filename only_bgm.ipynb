```python
df = df_raw
df
```

```python
df['댓글'] = df['댓글'].astype('string')
```

```python
df = df.dropna()
```

```python
comment_ = []
cnt = 0
cnt1 = 0

for i in range(len(df['댓글'])) :
    comment = df['댓글'][i].split()
    for word in comment :
        if '브금' in word :
            print(i)
            cnt += 1
            comment_.append(i)
        elif '노래' in word :
            cnt1 += 1
            comment_.append(i)
```

```python
bgm_df = df.loc[comment_]
bgm_df = bgm_df.drop_duplicates()
```

```python
bgm_df.to_excel("하루만네가되고싶어_bgm_최종.xlsx")
```
